## O que é Pesquisa Operacional?

A Pesquisa Operacional é uma disciplina que utiliza métodos analíticos avançados para auxiliar na tomada de decisões complexas. Ela combina matemática, estatística e computação para resolver problemas de otimização em diversas áreas, como logística, finanças, manufatura e telecomunicações.

![Representação de Programação Linear](img/pl.png)

## Programação Linear

A Programação Linear é uma técnica de otimização onde o objetivo é maximizar ou minimizar uma função linear, sujeita a um conjunto de restrições também lineares.

Um modelo de PL é composto por:
1. **Variáveis de Decisão**: O que queremos determinar (ex: quantidade de produtos a produzir).
2. **Função Objetivo**: O que queremos otimizar (ex: maximizar lucro ou minimizar custo).
3. **Restrições**: Limitações do sistema (ex: disponibilidade de matéria-prima, tempo de mão de obra).
4. **Restrições de Não-Negatividade**: Geralmente, as variáveis de decisão não podem ser negativas.

## Exemplo Prático: A Fábrica de Móveis

Imagine uma fábrica que produz dois tipos de móveis: Cadeiras e Mesas.

- Cada **Cadeira** gera um lucro de R$ 45,00.
- Cada **Mesa** gera um lucro de R$ 80,00.

Para a produção, temos as seguintes restrições de recursos:
1. **Madeira**: Estão disponíveis 400 unidades de madeira.
   - Uma cadeira consome 5 unidades.
   - Uma mesa consome 20 unidades.
2. **Mão de Obra**: Estão disponíveis 450 horas de trabalho.
   - Uma cadeira consome 10 horas.
   - Uma mesa consome 15 horas.

**Objetivo**: Quantas cadeiras e mesas a fábrica deve produzir para maximizar o lucro total?

### Formulação Matemática

Seja:
- $x_1$: quantidade de cadeiras
- $x_2$: quantidade de mesas

**Função Objetivo**:
Maximizar $Z = 45x_1 + 80x_2$

**Sujeito a**:
1. $5x_1 + 20x_2 \leq 400$ (Madeira)
2. $10x_1 + 15x_2 \leq 450$ (Mão de Obra)
3. $x_1, x_2 \geq 0$ (Não-negatividade)

## Implementação em Python com PuLP

Utilizaremos a biblioteca `PuLP`, que é uma ferramenta popular para modelagem de problemas de otimização linear em Python.

```python
# Instalando a biblioteca caso não esteja presente
# !pip install pulp

import pulp

# 1. Criar o problema de maximização
model = pulp.LpProblem("Maximizacao_de_Lucro", pulp.LpMaximize)

# 2. Definir as variáveis de decisão
x1 = pulp.LpVariable("Cadeiras", lowBound=0, cat='Continuous')
x2 = pulp.LpVariable("Mesas", lowBound=0, cat='Continuous')

# 3. Definir a Função Objetivo
model += 45 * x1 + 80 * x2, "Lucro_Total"

# 4. Definir as Restrições
model += 5 * x1 + 20 * x2 <= 400, "Restricao_Madeira"
model += 10 * x1 + 15 * x2 <= 450, "Restricao_Mao_de_Obra"

# 5. Resolver o problema
status = model.solve(pulp.PULP_CBC_CMD(msg=0))

# 6. Exibir os resultados
print(f"Status da Solução: {pulp.LpStatus[status]}")
print(f"Quantidade de Cadeiras (x1): {x1.varValue}")
print(f"Quantidade de Mesas (x2): {x2.varValue}")
print(f"Lucro Máximo Total: R$ {pulp.value(model.objective):.2f}")
```

## Conceitos Importantes para Computação

1. **Complexidade**: Muitos algoritmos de otimização linear, como o Simplex, têm complexidade exponencial no pior caso, mas são extremamente eficientes na prática. O algoritmo de Pontos Interiores oferece complexidade polinomial.
2. **Integração**: Problemas de PO são frequentemente partes de sistemas maiores. Por exemplo, um sistema de recomendação ou um roteador de pacotes pode usar otimização em tempo real.
3. **Programação Inteira**: Muitas vezes as variáveis devem ser inteiras (não podemos produzir meia cadeira). Nesses casos, o problema torna-se NP-difícil, exigindo técnicas como *Branch and Bound*.

## Problemas Clássicos em PO

Além da Programação Linear simples, a PO lida com problemas estruturais famosos:

1. **Problema da Mochila (Knapsack Problem)**: Dado um conjunto de itens com pesos e valores, determine quais itens incluir em uma mochila de capacidade limitada para maximizar o valor total.
2. **Problema do Caixeiro Viajante (TSP)**: Encontrar a rota mais curta que visita um conjunto de cidades e retorna à cidade de origem.

Estes problemas são fundamentais para entender a complexidade computacional e técnicas de heurísticas e meta-heurísticas.
